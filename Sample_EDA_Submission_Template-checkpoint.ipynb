## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a copy
df_clean = df.copy()

# Remove duplicates
df_clean.drop_duplicates(inplace=True)

# Fill numeric columns with median
numeric_cols = df_clean.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Fill categorical columns with mode
cat_cols = df_clean.select_dtypes(include=['object']).columns

for col in cat_cols:
    df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

print(df_clean.isnull().sum())
````

### What all manipulations have you done and insights you found?

**Data Manipulations Performed:**

1.Created a copy of the original dataset.

2.Checked for duplicate records and confirmed there were no duplicates.

3.Identified missing values in the dataset.

4.Filled missing values using the mode for categorical columns and the median for numerical columns.

5.Prepared a clean dataset for visualization and further analysis.

**Insights:**

1.The dataset contained no duplicate records.

2.Some columns had missing values, which were successfully handled.

3.The cleaned dataset is now ready for exploratory data analysis and machine learning.
